```python
def normalize(x, range=(0, 1)):
    return (x - min(x)) / (max(x) - min(x)) * (range[1] - range[0]) + range[0]
```

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr

def pearsonr_with_lag(x, y, lag):
    if lag > 0:
        x = x[lag:]
        y = y[:-lag]
    elif lag < 0:
        x = x[:lag]
        y = y[-lag:]
    
    return pearsonr(x, y)

def spearmanr_with_lag(x, y, lag):
    if lag > 0:
        x = x[lag:]
        y = y[:-lag]
    elif lag < 0:
        x = x[:lag]
        y = y[-lag:]
    
    return spearmanr(x, y)    
```

```python
def get_emotion_color(c):
    if c == 'Anger':
        return 'red'
    elif c == 'Disgust':
        return 'brown'
    elif c == 'Joy':
        return 'green'
    elif c == 'Sadness':
        return 'grey'
    elif c == 'Surprise':
        return 'blue'
    elif c == 'Fear':
        return 'orange'
```

## Some stats

```python
from scipy.stats import shapiro

# Check normality

#shapiro(aggregate_emotions_2020_df['Anger'])
for emotion in ['Fear', 'Anger', 'Sadness', 'Joy', 'Surprise', 'Disgust']:
    shap2019 = shapiro(aggregate_emotions_2019_df[emotion])
    shap2020 = shapiro(aggregate_emotions_2020_df[emotion])
    str_2019 = '2019 NORMAL'
    if shap2019[1] < 0.05:
        str_2019 = '2019 NOT NORMAL'
    str_2020 = '2020 NORMAL'
    if shap2020[1] < 0.05:
        str_2020 = '2020 NOT NORMAL'
    print(f'Shapiro-Wilk P-values for {emotion}: 2019 - {shap2019[1]:.3f}, 2020 - {shap2020[1]:.3f}, {str_2019}, {str_2020}')
```

```python
from scipy.stats import ttest_ind as ttest
from scipy.stats import mannwhitneyu

# for non-normal distributions run mann-whitney

print('Fear P-value:', mannwhitneyu(aggregate_emotions_2019_df['Fear'], aggregate_emotions_2020_df['Fear'])[1])
print('Anger P-value:', ttest(aggregate_emotions_2019_df['Anger'], aggregate_emotions_2020_df['Anger'], equal_var=False)[1]) # Welch's T-Test
print('Sadness P-value:', ttest(aggregate_emotions_2019_df['Sadness'], aggregate_emotions_2020_df['Sadness'], equal_var=False)[1]) # Welch's T-Test
print('Joy P-value:', ttest(aggregate_emotions_2019_df['Joy'], aggregate_emotions_2020_df['Joy'], equal_var=False)[1]) # Welch's T-Test
print('Surprise P-value:', mannwhitneyu(aggregate_emotions_2019_df['Surprise'], aggregate_emotions_2020_df['Surprise'])[1])
print('Disgust P-value:', mannwhitneyu(aggregate_emotions_2019_df['Disgust'], aggregate_emotions_2020_df['Disgust'])[1])
```

## Cross-correlation between all emotions 2019, 2020

```python
import os
import numpy as np
import json
import matplotlib.pyplot as plt

if not os.path.exists('correlation_data/between_emotions'):
    os.makedirs('correlation_data/between_emotions')

emotion_list = ['Anger', 'Disgust', 'Fear', 'Joy', 'Sadness', 'Surprise']
    
emotion_df_dict_2019 = {}
emotion_df_dict_2020 = {}
    
for year in [2019, 2020]:
    year_emotion_correlation_dict = {}
    if year == 2019:
        aggregate_emotions_df = aggregate_emotions_2019_df
        emotion_df_dict = emotion_df_dict_2019
    elif year == 2020:
        aggregate_emotions_df = aggregate_emotions_2020_df
        emotion_df_dict = emotion_df_dict_2020
        
    for i in range(len(emotion_list)):
        for j in range(i + 1, len(emotion_list)):
            emotion1 = emotion_list[i]
            emotion2 = emotion_list[j]
            
            if (emotion1, emotion2) not in emotion_df_dict:
                emotion_df_dict[(emotion1, emotion2)] = []
            
            xcorr = np.correlate(normalize(aggregate_emotions_df[emotion1]), normalize(aggregate_emotions_df[emotion2]), 'full')
            np.save('correlation_data/between_emotions/xcorr_{}_{}_{}.npy'.format(emotion1, emotion2, year), xcorr)

            fig = plt.figure(figsize=(22, 10))
            plt.plot(xcorr, '-o')
            plt.title('Cross-correlation between {} and {} in {}'.format(emotion1, emotion2, year))
            plt.xticks(list(range(len(xcorr))), labels=[x - len(aggregate_emotions_df) + 1 for x in range(len(xcorr))])
            plt.tight_layout()
            plt.savefig('correlation_data/between_emotions/xcorr_{}_{}_{}.png'.format(emotion1, emotion2, year))

            # dont display it here
            plt.close(fig)
            
            highest_xcorr_lag = np.argmax(xcorr) - len(aggregate_emotions_df) + 1
            highest_xcorr_pearsonr = pearsonr_with_lag(aggregate_emotions_df[emotion1], aggregate_emotions_df[emotion2], highest_xcorr_lag)
            result_dict = {'highest_xcorr_lag': str(highest_xcorr_lag), 
                           'highest_xcorr_pearonsr_corr': str(highest_xcorr_pearsonr[0]),
                            'highest_xcorr_pearsonr_pvalue': str(highest_xcorr_pearsonr[1])}
            year_emotion_correlation_dict['{}-{}'.format(emotion1, emotion2)] = result_dict
            emotion_df_dict[(emotion1, emotion2)] = highest_xcorr_pearsonr
            
            fig = plt.figure(figsize=(15, 8))
            plt.plot(normalize(aggregate_emotions_df[emotion1]), color=get_emotion_color(emotion1))
            plt.plot(normalize(aggregate_emotions_df[emotion2]), color=get_emotion_color(emotion2))
            plt.title('Normalized {} and {} in {}'.format(emotion1, emotion2, year))
            plt.legend([emotion1, emotion2])
            plt.xticks(list(range(len(aggregate_emotions_df))), ['{} to {}'.format(date1[:date1.index(' ')], date2[:date2.index(' ')]) for (idx, date1, date2) in
                                                                     aggregate_emotions_df[['start_date', 'end_date']].itertuples()], rotation=90)
            plt.tight_layout()

            plt.savefig('correlation_data/between_emotions/normalized_{}_{}_{}.png'.format(emotion1, emotion2, year))
            
            # dont display it here
            plt.close(fig)
            
    with open('correlation_data/between_emotions/highest_xcorr_{}.json'.format(year), 'w') as f:
        json.dump(year_emotion_correlation_dict, f)
```

```python
df_dict_2019 = {'Emotion': emotion_list, 'Anger': [], 'Disgust': [], 'Fear': [], 'Joy': [], 'Sadness': [], 'Surprise': []}
df_dict_2020 = {'Emotion': emotion_list, 'Anger': [], 'Disgust': [], 'Fear': [], 'Joy': [], 'Sadness': [], 'Surprise': []}

for year in [2019, 2020]:
    if year == 2019:
        emotion_df_dict = emotion_df_dict_2019
        df_dict = df_dict_2019
    elif year == 2020:
        emotion_df_dict = emotion_df_dict_2020
        df_dict = df_dict_2020
        
    temp = {}
        
    for (emotion1, emotion2) in emotion_df_dict:
        temp[(emotion2, emotion1)] = emotion_df_dict[(emotion1, emotion2)]
    
    for tup in temp:
        emotion_df_dict[tup] = temp[tup]

    for (emotion1, emotion2) in emotion_df_dict:
        df_dict[emotion1].append((emotion2, emotion_df_dict[(emotion1, emotion2)]))

    for emotion in emotion_list:
        df_dict[emotion] = [(emotion_list.index(em), tup) for (em, tup) in df_dict[emotion]]
        df_dict[emotion].append((emotion_list.index(emotion), (1.0, 0.0)))
```

```python
for emotion in emotion_list:
    df_dict_2019[emotion] = sorted(df_dict_2019[emotion], key=lambda tup: tup[0])
    df_dict_2020[emotion] = sorted(df_dict_2020[emotion], key=lambda tup: tup[0])
```

```python
import copy

df_dict_2019_pearson_corr = copy.deepcopy(df_dict_2019)
df_dict_2020_pearson_corr = copy.deepcopy(df_dict_2020)
df_dict_2019_pearson_pvalue = copy.deepcopy(df_dict_2019)
df_dict_2020_pearson_pvalue = copy.deepcopy(df_dict_2020)
for emotion in emotion_list:
    df_dict_2019_pearson_corr[emotion] = [corr for (idx, (corr, pvalue)) in df_dict_2019_pearson_corr[emotion]]
    df_dict_2020_pearson_corr[emotion] = [corr for (idx, (corr, pvalue)) in df_dict_2020_pearson_corr[emotion]]
    
    df_dict_2019_pearson_pvalue[emotion] = [pvalue for (idx, (corr, pvalue)) in df_dict_2019_pearson_pvalue[emotion]]
    df_dict_2020_pearson_pvalue[emotion] = [pvalue for (idx, (corr, pvalue)) in df_dict_2020_pearson_pvalue[emotion]]
```

```python
def get_emotion_color(c):
    if c == 'Anger':
        return 'red'
    elif c == 'Disgust':
        return 'brown'
    elif c == 'Joy':
        return 'green'
    elif c == 'Sadness':
        return 'grey'
    elif c == 'Surprise':
        return 'blue'
    elif c == 'Fear':
        return 'orange'
```

```python
emotion = 'Disgust'

def get_list_elements_with_idx_mod(lst, mod):
    new_lst = []
    for i, el in enumerate(lst):
        if i % mod == 0:
            new_lst.append(el)
    return new_lst

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rc('font', **{'family': 'normal', 'weight': 'normal', 'size': 12})

colors = ['red', 'brown', 'orange', 'limegreen', 'grey', 'deepskyblue']


ax = aggregate_emotions_2020_US_df[emotion].plot(color=get_emotion_color(emotion), figsize=(13, 5))
ax.plot(aggregate_emotions_2020_UK_df[emotion], color=get_emotion_color(emotion), ls='--')
ax.plot(aggregate_emotions_2020_Canada_df[emotion], color=get_emotion_color(emotion), ls=':')
plt.gcf().subplots_adjust(bottom=0.3)
#plt.ylim(min(aggregate_emotions_2019_df[emotion].min(), aggregate_emotions_2020_df[emotion].min()), max(aggregate_emotions_2019_df[emotion].max(), aggregate_emotions_2020_df[emotion].max()))
ax.set_xticks(get_list_elements_with_idx_mod(aggregate_emotions_2020_US_df.index, 4))
ax.set_xticklabels(get_list_elements_with_idx_mod([entry[1][entry[1].index('-')+1:entry[1].index(' ')] 
                                       for entry in aggregate_emotions_2020_US_df[['start_date', 'end_date']].itertuples()], 4), rotation=90)
#ax.set_xticklabels([])
plt.legend(['{}, US'.format(emotion), '{}, UK'.format(emotion), '{}, Canada'.format(emotion)])
#plt.savefig('{}_2020_Canada.png'.format(emotion))
plt.show()
```
